```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/content/E-commerce.csv")
df.info()
df.info(verbose=True)
print(df.columns)
sns.pairplot(df)
plt.show()


df['Product Reviews'] = pd.to_numeric(df['Product Reviews'], errors='coerce')

df['Product Reviews'].plot.hist(bins=25, figsize=(8, 4))
plt.show()
plt.figure(figsize=(10, 7))


numeric_df = df.select_dtypes(include=np.number)
sns.heatmap(numeric_df.corr(), annot=True, linewidths=2)
plt.show()
df.describe()
df.corr
plt.figure(figsize=(10, 5))
plt.ylabel('Frequency')
plt.show()
df.head()
df.tail()
df.shape
df.size

average_income_by_gender = df.groupby('Gender')['Annual Income'].mean
print(average_income_by_gender)
```
