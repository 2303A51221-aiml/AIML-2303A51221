Lab task is to visualize the data set or the data using matplotlib


```python
import matplotlib.pyplot as plt

people = ["Sai sathwik","Adithya","Gama","Sushruth"]
age = [18,18,17,19]
weight = [75,49,60,69]
height = [177,180,168,160]

plt.scatter(weight,height)
plt.title("Relationship between age and weight")
plt.ylabel("height in centimeters")
plt.xlabel("weight in kgs")

plt.show()

plt.scatter(age,weight)
plt.title("Relationship between age and weight")
plt.ylabel("weight in kgs")
plt.xlabel("age")
plt.show()
```

### Generate bar plot

```python
plt.figure(figsize=(12,4))
plt.title("peoples weight in kgs")
plt.bar(x=people,height=weight,width=0.6,alpha = 0.6,color = "red")
plt.xlabel("peoples",fontsize = 15)
plt.ylabel("weight (in kgs)",fontsize=15)
plt.show()







```

Generate histogram

```python
plt.figure(figsize=(7,5))

plt.hist(weight,color = "orange",edgecolor = "k",alpha = 0.75,bins = 5)

plt.title("Histogram of patient weight",fontsize = 18)
plt.xlabel("weight in kgs",fontsize = 18)
plt.show()
```
